```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
x=cancer.data
y=cancer.target
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y, stratify=y, test_size=0.2, random_state=42)
```

```python
# 단일 신경망 구현
class SingleLayer:
    def __init__(self):
        self.w=None
        self.b=None
        self.losses=[]

    def forpass(self, x):
        z=np.sum(x*self.w)+self.b #np.sum은 배열 모든 요소의 합을 계산한다. 
        return z
    
    def backprop(self, x, err):
        w_grad=x*err
        b_grad=1*err
        return w_grad, b_grad
        #활성화 함수 구현(시그모이드)
    
    def activation(self, z):
        z=np.clip(z, -100, None) # clip 함수는 list, min, max로 사용한다. 
        a=1/(1+np.exp(-z))
        return a   

    def fit(self, x, y, epochs=100):
        self.w=np.ones(x.shape[1]) #x.shape가 Tuple의 자료형이기 때문에 뽑는 것이다 열 정보를 뽑은 것임. 
                                    #즉, 열만큼의 1 배열을 만들어서 초깃값을 설정해 주고 있다. 
        self.b=0
        for i in range(epochs):
            loss=0
            indexes=np.random.permutation(np.arange(len(x))) 
            # random.permutation은 배열을 받으면 무작위로 섞어준다. 정수를 받으면 그만큼의 랜덤 배열 생성한다.
            # arrange는 들어온 인수만큼의 배열을 생성한다. 따라서 0~29의 index를 만들고 섞는 것이다. 

            for i in indexes: #zip은 서로 다른 자료형을 묶어 주는 함수이다. 
                z=self.forpass(x[i])
                a=self.activation(z)
                err=y[i]-a
                w_grad, b_grad=self.backprop(x[i], err)
                self.w+=w_grad
                self.b+=b_grad
                a=np.clip(a, 1e-10, 1-1e-10)
                loss+=y[i]*np.log(a)+(1-y[i])*np.log(1-a)
            self.losses.append(loss/len(y))

    def predict(self, x): #모델 예측 함수 구현
        z=[self.forpass(x_i) for x_i in x] #리스트 내포 문법. 
        return np.array(z)>0

    def score(self, x, y): # 정확도 계산 메커니즘. 모델 정확도를 평가한다. 
        return np.mean(self.predict(x)==y)
```

```python
# 단일 신경망 훈련

layer=SingleLayer()
layer.fit(x_train, y_train)
layer.score(x_test, y_test)

#plt.plot은 기본 배열을 y값으로 그래프를 그려 준다. 
```
